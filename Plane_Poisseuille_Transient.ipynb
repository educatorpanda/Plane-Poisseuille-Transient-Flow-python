## Plane Poisseuille Transient Flow

### Time Developing Flow in a Parallel Plate Channel

Consider the time evolution of the flow of a fluid (initially at rest) between two parallel
plates separated by a distance $2h$. The governing Navier-Stokes equations for this flow are
simplified to:

$$\frac{\partial u}{\partial t} = -\frac{1}{\rho }\frac{\mathrm{d} p}{\mathrm{d} x}+\nu \frac{\partial^2 u}{\partial y^2} $$

where $ \large \frac{\mathrm{d} p}{\mathrm{d} x}$ is a constant pressure gradient.

<img src="TPPF_schematic.PNG">

**Nondimensionalize all the variables and the governing equation (including initial and boundary conditions)**

Let $$ -\frac{1}{\rho }\frac{\mathrm{d} p}{\mathrm{d} x} = s = \text{constant}$$
$$ \therefore \frac{\partial u}{\partial t} = s+\nu \frac{\partial^2 u}{\partial y^2}  $$ 

Let Non-dimensional variables: $\large U = \frac{\nu u}{4sh^2}$, $\large Y = \frac{y}{2h}$, $\large T = \frac{\nu t}{4h^2}$

Therefore Non-dimensional equation will be:

$$ \frac{\partial U}{\partial T} = 1 + \frac{\partial^2 U}{\partial Y^2} \hspace{0.5cm} ....(1)$$

**Initial and Boundary conditions in Dimensional form:**
$$u(y,0)=0$$
$$u(0,t)=0$$
$$u(2h,t)=0$$ 

**Initial and Boundary conditions in Non-Dimensional form:**
$$u(Y,0)=0$$
$$u(0,T)=0$$
$$u(1,T)=0$$ 

**Determine the ‘exact’ solution**

Let $$ U(Y,T) = U_{1}(Y,T)+U_{2}(Y,T)$$
Substitute it in $(1)$ to get
$$ \frac{\partial U_1}{\partial T} + \frac{\partial U_2}{\partial T} = 1 + \frac{\partial^2 U_1}{\partial Y^2} + \frac{\partial^2 U_2}{\partial Y^2} $$
<br>
Let $U_1$ represents a steady state solution: 
<br>
$$\therefore \frac{\partial U_1}{\partial T} = 0 \hspace{0.5cm} \text{and} \hspace{0.5cm} 1 + \frac{\partial^2 U_1}{\partial Y^2} = 0 \hspace{0.5cm} ...(2)$$
<br>
i.e. $U_1(Y,T)\equiv U_1(Y)$. Also for $U_2$:
<br>
$$ \frac{\partial U_2}{\partial T} = \frac{\partial^2 U_2}{\partial Y^2}  \hspace{0.5cm} ...(3)$$
<br>
Solving for $U_1$ using $(2)$ and Boundary conditions gives:
<br>
$$U_1 = \frac{1}{2}Y(1-Y)$$
<br>
B.C. for $U_2$:
$$U(Y,0)=0=U_{1}(Y,0)+U_{2}(Y,0)$$
$$\therefore U_{2}(Y,0) = -U_{1}(Y)$$
<br>
Also, $$U_{2}(0,T)=U_{2}(1,T)=0$$
<br>
Therefore from $(3)$ and the above B.C. (using separation of variables method) we get
<br>
<br>
$$U_{2}(Y,T) =  -\sum_{n=0}^{\infty }\left [ \frac{4}{((2n+1)\pi )^3}\text{sin}((2n+1)\pi ) e^{-\lambda _{n}^{2}T}\right ]; \hspace{1cm} \text{where} \hspace{0.3cm} \lambda _{n} = (2n+1)\pi$$
<br>
$$\therefore U(Y,T) = \frac{1}{2}Y(1-Y) -\sum_{n=0}^{\infty }\left [ \frac{4}{((2n+1)\pi )^3}\text{sin}((2n+1)\pi ) e^{-\lambda _{n}^{2}T}\right ]$$

**Flow Rate** $Q(T)$

$$ Q(T) = \int_{0}^{1}U(Y,T)dY $$
<br>
$$\therefore Q(T) = \frac{1}{12} -\sum_{n=0}^{\infty }\left [ \frac{8}{((2n+1)\pi )^4}e^{-\lambda _{n}^{2}T}\right ]$$

**Shear Stress** $\tau (T)$

$$\large \tau (T) = \left.\begin{matrix}
\frac{\partial U}{\partial Y}
\end{matrix}\right|_{Y=0} $$
<br>
$$\therefore \tau (T) = \frac{1}{2} -\sum_{n=0}^{\infty }\left [ \frac{4}{((2n+1)\pi )^2}e^{-\lambda _{n}^{2}T}\right ]$$
